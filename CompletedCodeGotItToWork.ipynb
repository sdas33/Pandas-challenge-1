```python
# Import necessary libraries
import pandas as pd
import numpy as np
import os

# Define the relative file path
file_path = os.path.join('M4_Starter_Code', 'client_dataset (1).csv')

# Verify the file exists before proceeding
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found at: {file_path}")

# Load the dataset
df = pd.read_csv(file_path)
print("Dataset loaded successfully!")
print(df.head())

# Part 1: Explore the Data
# 1. View the column names
print("\nColumn Names:")
print(df.columns)

# 2. Use the describe function to gather basic statistics
print("\nDataset Statistics:")
print(df.describe())

# 3. What three item categories had the most entries?
top_categories = df['category'].value_counts().head(3)
print("\nTop 3 Item Categories:")
print(top_categories)

# 4. For the category with the most entries, which subcategory had the most entries?
most_common_category = top_categories.idxmax()
top_subcategory = df[df['category'] == most_common_category]['subcategory'].value_counts().idxmax()
print(f"\nTop Subcategory in {most_common_category}: {top_subcategory}")

# 5. Which five clients had the most entries in the data?
top_clients = df['client_id'].value_counts().head(5)
print("\nTop 5 Clients by Entries:")
print(top_clients)

# 6. Store the client IDs of the top 5 clients in a list
top_client_ids = top_clients.index.tolist()
print("\nTop 5 Client IDs:")
print(top_client_ids)

# 7. How many total units (qty) did the client with the most entries order?
top_client_total_units = df[df['client_id'] == top_client_ids[0]]['qty'].sum()
print(f"\nTotal Units Ordered by Client {top_client_ids[0]}: {top_client_total_units}")

# Part 2: Transform the Data
# 1. Create a column that calculates the subtotal for each line
df['subtotal'] = df['qty'] * df['unit_price']

# 2. Create a column for shipping price
df['shipping_price'] = np.where(df['weight'] > 50, df['weight'] * 7, df['weight'] * 10)

# 3. Create a column for the total price with 9.25% sales tax
df['total_price'] = (df['subtotal'] + df['shipping_price']) * 1.0925

# 4. Create a column for the cost of each line
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']

# 5. Create a column for the profit of each line
df['profit'] = df['total_price'] - df['line_cost']

# Part 3: Confirm Your Work
# Confirm the total prices for the provided Order IDs
order_totals = {
    2742071: 152811.89,
    2173913: 162388.71,
    6128929: 923441.25
}

print("\nConfirming Order Totals:")
for order_id, expected_total in order_totals.items():
    calculated_total = df[df['order_id'] == order_id]['total_price'].sum()
    print(f"Order ID {order_id} - Expected: {expected_total}, Calculated: {calculated_total:.2f}")

# Part 4: Summarize and Analyze
# Calculate total revenue, units, shipping, and profit for the top 5 clients
summary = df[df['client_id'].isin(top_client_ids)].groupby('client_id').agg(
    total_units=('qty', 'sum'),
    total_shipping=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()

# Convert revenue and profit to millions
summary['total_revenue'] = summary['total_revenue'] / 1_000_000
summary['total_profit'] = summary['total_profit'] / 1_000_000

# Rename columns for presentation
summary.rename(columns={
    'total_units': 'Total Units Purchased',
    'total_shipping': 'Total Shipping Cost',
    'total_revenue': 'Total Revenue (Millions)',
    'total_profit': 'Total Profit (Millions)'
}, inplace=True)

# Sort by Total Profit
summary = summary.sort_values(by='Total Profit (Millions)', ascending=False)

print("\nSummary of Top 5 Clients:")
print(summary)

# Part 4: Brief Summary
print("\nFindings:")
print("1. The top 5 clients by quantity purchased contributed significantly to revenue and profit.")
print("2. The client with the highest profit demonstrated strong revenue contributions.")
print("3. Shipping costs and weights significantly impact overall profitability.")
```
